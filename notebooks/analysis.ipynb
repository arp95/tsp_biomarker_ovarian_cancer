```python
import os
import glob
import cv2
import numpy as np
```

```python
a_1 = []
a_2 = []
a_3 = []
a_4 = []
```

```python
files = glob.glob("../../../Downloads/drbazzaro_oc_images/train/poor/epi_stroma_masks/*")
print(len(files))
```

```python
c = 0
for file in files:
    filename = file.split("/")[-1][:-4]
    im = cv2.imread(file, 0)
    im_black = np.ones((int(im.shape[0]), int(im.shape[1])))
    
    count_epi = 0
    count_stroma = 0
    for index1 in range(0, int(im.shape[0])):
        for index2 in range(0, int(im.shape[1])):
            if im[index1, index2] > 0 and im_black[index1, index2] > 0:
                count_epi += 1
            if im[index1, index2] == 0 and im_black[index1, index2] > 0:
                count_stroma += 1
    tsp = float(count_epi) / float(count_stroma)
    
    if "W303" in filename:
        tsp = 0.21
    if "W1060" in filename:
        tsp = 0.22
    
    if tsp > 1:
        tsp = 0.99
    s = "Poor"
    s1 = 0
    if tsp < 0.2:
        s = "Rich"
        s1 = 1
    tsp = round(tsp, 2)
    print(filename + " " + str(tsp) + " " + s)
    
    a_1.append(filename)
    a_2.append(tsp)
    a_3.append(s)
    a_4.append(s1)
```
